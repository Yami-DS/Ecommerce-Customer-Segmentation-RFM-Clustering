```python
# Segmenting customers based on Recency(Date), Frequency(InvoiceNo) and Monetary(Price)
```

```python
RFMScore = dataset.groupby('CustomerID').agg({'Date':lambda x: (Latest_date - x.max()).days,'InvoieNo': lambda x: x.count(), 'Price': lambda x: x.sum()})
```

```python
RFMScore
```

```python
RFMScore.rename(columns={'Date':'Recency','InvoieNo':'Frequency','Price':'Monetary'}, inplace = True)
```

```python
RFMScore
```

```python
# Model Building
```

```python
RFMScore.Recency.describe()
```

```python
# split the data into quantile method

quantiles = RFMScore.quantile(q=[0.25,0.50,0.75])
quantiles = quantiles.to_dict()
quantiles
```

```python
# Recency output explanation:

# 25% customers are 10 days old
# 50% customers are 19 days old
# 75% customers are 73 days old
```

```python
# here 25% we're considering because 25% customer visited shop within 10 days i.e. very recent
def RecencyScore(x,p,d):  # x:dataset,p:percentage.d:datapoint 
    if x<=d[p][0.25]:
        return 1
    elif x<=d[p][0.50]:
        return 2
    elif x<=d[p][0.75]:
        return 3
    else:
        return 4
```

```python
RFMScore.Frequency.describe()
```

```python
# In Frequency the max is 7970 so 7970 a customer is visited hence in this case we consider 75% and above as higher
```

```python
def FreqMonetaryScore(x,p,d):
    if x<=d[p][0.25]:
        return 4
    elif x<=d[p][0.50]:
        return 3
    elif x<=d[p][0.75]:
        return 2
    else:
        return 1
```

```python
RFMScore.Monetary.describe()
```

```python
RFMScore.columns
```

```python
RFMScore['R'] = RFMScore['Recency'].apply(RecencyScore,args=('Recency',quantiles))
RFMScore['F'] = RFMScore['Frequency'].apply(FreqMonetaryScore,args=('Frequency',quantiles))
RFMScore['M'] = RFMScore['Monetary'].apply(FreqMonetaryScore,args=('Monetary',quantiles))
```

```python
RFMScore.reset_index()
```

```python
RFMScore['RFMValue'] = RFMScore[['R','F','M']].sum(axis=1)
```

```python
RFMScore
```

```python
RFMScore['RFMGroup'] = RFMScore.R.map(str) + RFMScore.F.map(str) + RFMScore.M.map(str)
```

```python
RFMScore.reset_index()
```

```python
Loyalty_Level = ['Platinum','Diamond','Gold','Silver']
score_cuts = pd.qcut(RFMScore.RFMValue,q=4,labels =Loyalty_Level ) # qcut is quantilecut an inbuilt function
RFMScore['Loyalty_Level'] = score_cuts.values
RFMScore = RFMScore.reset_index()
RFMScore
```

```python
RFMScore.to_csv('final_data.csv')
```
